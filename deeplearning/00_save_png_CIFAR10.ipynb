```python
import pickle
import os
import png # pypng library
import numpy as np
```

```python
file_path = os.getcwd()
file_path
```

```python
lp = list(map('data_batch_{}'.format, range(1,2)))
lp.append('test_batch')

for batch in lp:
    fpath = os.path.join(file_path,'cifar-10-batches-py',batch)

    f = open(fpath, 'rb')
    labels = pickle.load(open(os.path.join(file_path,'cifar-10-batches-py', 'batches.meta'), 'rb'), encoding="ASCII")

    d = pickle.load(f, encoding='bytes')

    # decode utf8
    d_decoded = {}
    for k, v in d.items():
        d_decoded[k.decode('utf8')] = v

    d = d_decoded
    f.close()
  
    if batch in 'test_batch':
        for i, filename in enumerate(d['filenames']):
            folder = os.path.join(
                file_path,
                'CIFAR10',
                'test',
                labels['label_names'][d['labels'][i]]
            )
            os.makedirs(folder, exist_ok=True)
            q = d['data'][i]
            print(filename)
            with open(os.path.join(folder, filename.decode()), 'wb') as outfile:
                png.from_array(q.reshape((32, 32, 3), order='F').swapaxes(0,1), mode='RGB').save(outfile)
    else:
        for i, filename in enumerate(d['filenames']):
            folder = os.path.join(
                file_path,
                'CIFAR10',
                'train',
                labels['label_names'][d['labels'][i]]
            )
            os.makedirs(folder, exist_ok=True)
            q = d['data'][i]
            print(filename)
            with open(os.path.join(folder, filename.decode()), 'wb') as outfile:
                png.from_array(q.reshape((32, 32, 3), order='F').swapaxes(0,1), mode='RGB').save(outfile)
            
```
